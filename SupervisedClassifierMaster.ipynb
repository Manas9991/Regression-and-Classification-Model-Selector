##Importing Libraries:

```
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

#Loading Data:

```
data_address = r'employee_data.csv' #Pass address of data set here
df = pd.read_csv(data_address)
```

##Data Cleaning:

It is recommended to modify this code for every project after making a copy of this code. This is just a non working sample.

```
df.head()
```

```
df['salary'] = df['salary'].apply(lambda x: 0 if x=='low' else 1)
```

```
target = 'quit' #Enter Target Variable column name here
y = df[target]
X = df.drop(columns = [target, 'department'])
```

##Applying ML Models:

```
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, jaccard_score
models = ['Logistic Regession', 'KNN', 'SVM', 'GaussianNB', 'Random Forest', 'Gradient Boosting']
accuracy = []
f1 = []
precision = []
recall = []
jaccard = []
```

```
def evaluate(truth, pred):
  accuracy.append(accuracy_score(truth, pred))
  f1.append(f1_score(truth, pred))
  precision.append(precision_score(truth, pred))
  recall.append(recall_score(truth, pred))
  jaccard.append(jaccard_score(truth, pred))
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8)
```

```
from sklearn.preprocessing import Normalizer
norm = Normalizer().fit(X_train)
X_train = norm.transform(X_train)
X_test = norm.transform(X_test)
```

```
LoR = LogisticRegression()
LoR.fit(X_train, y_train)
y_pred = LoR.predict(X_test)
evaluate(y_test, y_pred)
```

```
t = 0
K = 0
for i in   range(1, 11):
  KNN = KNeighborsClassifier(n_neighbors=i)
  KNN.fit(X_train, y_train)
  y_pred = KNN.predict(X_test)
  if(t<=accuracy_score(y_test, y_pred)*100):
    t = accuracy_score(y_test, y_pred)*100
    K = i
KNN = KNeighborsClassifier(n_neighbors=K)
KNN.fit(X_train, y_train)
y_pred = KNN.predict(X_test)
evaluate(y_test, y_pred)
```

```
SVM = SVC(kernel = 'rbf') #select according to application
SVM.fit(X_train, y_train)
y_pred = SVM.predict(X_test)
evaluate(y_test, y_pred)
```

```
GNB = GaussianNB()
GNB.fit(X_train, y_train)
y_pred = GNB.predict(X_test)
evaluate(y_test, y_pred)
```

```
#Select n_estimators and depth according to application
RF = RandomForestClassifier(n_estimators = 100, criterion='gini')
RF.fit(X_train, y_train)
y_pred = RF.predict(X_test)
evaluate(y_test, y_pred)
```

```
GBC = GradientBoostingClassifier()
GBC.fit(X_train, y_train)
y_pred = GBC.predict(X_test)
evaluate(y_test, y_pred)
```

```
Final = {'Model': models, 'Accuracy': accuracy, 'F1 score': f1, 'Jaccard score' : jaccard, 'Precision': precision, 'Recall': recall}
Final = pd.DataFrame(data=Final)
Final
```
